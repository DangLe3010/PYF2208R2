```python
'''
Bài 4: Nhập vào số nguyên dương n. Tính tổng các chữ số của n.
'''
```

```python
def totalDigitsOfNumber(n):
    total = 0;
    while (n > 0):
        total = total + n % 10;
        n = int(n / 10);
    return total;
 
n = int(input("Nhập số nguyên dương n = "));
print("Tổng các chữ số của", n , "là", totalDigitsOfNumber(n));
```

```python
'''
Bài 7: Nhập vào một số nguyên dương n. Tính giai thừa của n.
Note: Giai thừa của n được tính bằng công thức: n! = n*(n - 1)*(n - 2)*...*1
Ví dụ: 5! = 5 × 4 × 3 × 2 × 1 = 120
'''
```

```python
print("Nhập giá trị n: ",end='')
n=int(input())
if n > 0:
    giaithua=1
    for i in range(1,n+1):
        giaithua=giaithua*i
    print(n,"giai thừa bằng:",giaithua)
else:
    print("Vui lòng nhập n > 0")
```

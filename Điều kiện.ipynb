## Câu lệnh if

```python
mark = float(input("nhập điểm cuối kì: "))
if mark >= 8.5:
    print("bạn được tuyển thẳng")
```

```python
mark = float(input("nhập điểm cuối kì: "))
if mark >= 8.5:
    print("bạn được tuyển thẳng")
else:
    print("bạn chưa đủ điểm")
```

## Câu lệnh if..else

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
num = int(input("nhập số nguyên: "))
if num % 2 == 0:
    print(f"{num} la so chan")
else:
    print(f"{num} la so le")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
string = input("nhap chuoi: ")
if "python" in string:
    print(f"'{string}' co chua cum tu 'python'")
else:
    print(f"'{string}' khong chua cum tu 'python'")
```
